```python
import requests
import csv
import json
from collections import defaultdict

# CSV 파일 경로
csv_file_path = "C:\\Users\user\\Documents\\카카오톡 받은 파일\\refer.csv"

# CSV 파일에서 contentId, contentTypeId 값을 읽어오는 함수
def read_csv(file_path):
    content_id_list = []
    content_type_id_list = []
    try:
        with open(file_path, newline='', encoding='utf-8-sig') as csvfile:
            reader = csv.reader(csvfile)
            next(reader)  # 첫 번째 줄은 헤더이므로 건너뜁니다.
            for row in reader:
                content_id_list.append(row[0])  # A열 (contentId)
                content_type_id_list.append(row[1])  # B열 (contentTypeId)
    except FileNotFoundError:
        print(f"CSV 파일을 찾을 수 없습니다: {file_path}")
        return [], []
    except Exception as e:
        print(f"CSV 파일 읽기 중 오류 발생: {e}")
        return [], []
    return content_id_list, content_type_id_list

# API 요청을 보낼 URL을 만드는 함수
def get_api_url(content_id, content_type_id):
    return f"https://apis.data.go.kr/B551011/KorService1/detailIntro1?serviceKey=uBSmr7z03f20VpfRoX%2BKCs%2B70B4BgwEVZdGTWrnanJ4jX9qdyqvbPTPER2Fc9P7Ku5Bh%2FsQvLQvEu7ldiDP6yA%3D%3D&MobileOS=ETC&MobileApp=AppTest&_type=json&contentId={content_id}&contentTypeId={content_type_id}&numOfRows=10&pageNo=1"

# 오류가 발생한 ID를 기록할 리스트
error_log = []

# 각 contentTypeId 별 데이터 저장을 위한 defaultdict
data_by_content_type = defaultdict(list)

# CSV 파일에서 코드 읽어오기
content_id_list, content_type_id_list = read_csv(csv_file_path)

# 각 contentId, contentTypeId에 대해 API 요청 보내기
for content_id, content_type_id in zip(content_id_list, content_type_id_list):
    try:
        # URL 생성
        url = get_api_url(content_id, content_type_id)

        # API 요청 보내기
        response = requests.get(url)

        # 요청 성공 여부 확인
        if response.status_code == 200:
            # JSON 데이터 받기
            data = response.json()

            # 결과 출력 (필요한 데이터만 출력)
            print(json.dumps(data, indent=4, ensure_ascii=False))  # 보기 쉽게 출력

            # 데이터를 CSV로 저장
            all_data = data.get('response', {}).get('body', {}).get('items', {}).get('item', [])

            # contentTypeId 별로 데이터 저장
            if all_data:
                data_by_content_type[content_type_id].extend(all_data)

                print(f"contentId: {content_id}, contentTypeId: {content_type_id} 데이터가 저장되었습니다.")
        else:
            raise Exception(f"요청 실패. 상태 코드: {response.status_code}")
    except Exception as e:
        print(f"contentId: {content_id}, contentTypeId: {content_type_id} 처리 중 오류 발생: {e}")
        error_log.append((content_id, content_type_id))

# contentTypeId 별로 데이터를 파일에 저장
for content_type_id, records in data_by_content_type.items():
    file_name = f"output_{content_type_id}.csv"
    with open(file_name, "w", newline="", encoding="utf-8-sig") as csvfile:
        if records:
            fieldnames = records[0].keys()
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(records)
    print(f"contentTypeId: {content_type_id} 데이터가 {file_name} 파일로 저장되었습니다.")

# 오류 로그 출력
if error_log:
    print("\n오류가 발생한 ID 목록:")
    for error_id, error_type in error_log:
        print(f"contentId: {error_id}, contentTypeId: {error_type}")

    # 오류 로그를 파일로 저장
    with open("error_log.csv", "w", newline="", encoding="utf-8-sig") as error_file:
        writer = csv.writer(error_file)
        writer.writerow(["contentId", "contentTypeId"])
        writer.writerows(error_log)

    print("오류 로그가 error_log.csv 파일에 저장되었습니다.")
```
